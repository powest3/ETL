```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#import CSV into Pandas
csv_file = "../Desktop/winemag2.csv"
wine_rev_df = pd.read_csv(csv_file)
wine_rev_df.head()
```

```python
#Create new dataframe using columns we choose
new_wine_df = wine_rev_df[['country', 'points', 'price', 'province', 'region_1', 'region_2', 'variety', 'winery']].copy()
new_wine_df.head()
```

```python
wine_rev_df.columns
```

```python
#Drop Nulls
new_wine_df.dropna()
```

```python
#Enter in Serial Key(Postgres) to overcome null, connect to Db
rds_connection_string = "postgres:Blacks21@localhost:5432/Wine_Db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#Check tables
engine.table_names()
```

```python
#Load CSV/converted Dataframe to Postgres
new_wine_df.to_sql(name='wine_reviews', con=engine, if_exists='append', index=False)
```
